```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
qc = QuantumCircuit(2)

qc.h(0)
phi_1 = 2* np.arccos(1 / np.sqrt(30))
qc.rz(phi_1, 0)

qc.cx(0, 1)

qc.rz(phi_1, 0)

qc.draw('mpl')
```

```python
# Simulate the circuit
state = Statevector.from_instruction(qc)
s = state.draw('latex')
s
```

```python
qc.h(0)
qc.h(1)
state = Statevector.from_instruction(qc)

s = state.draw('latex')
s
```

```python
from qiskit.quantum_info import Operator
qc2 = QuantumCircuit(2)
qc2.h(0)
qc2.h(1)
qc2.z(0)
qc2.z(1)
qc2.h(0)
qc2.h(1)

qc2.draw('mpl')
```

```python
# get matrix fro circuit
op = Operator(qc2)
op.draw('latex')
```

```python
qc3 = QuantumCircuit(2)

a0 = 1 / np.sqrt(30)
a1 = 2 / np.sqrt(30)
a3 = 3 / np.sqrt(30)
a4 = 4 / np.sqrt(30)

qc3.h(0)
qc3.h(1)

def controlled_rz(qc, control, target, angle):
    # from CNOT and Rz
    qc.cx(control, target)
    qc.rz(angle / 2, target)
    qc.cx(control, target)
    qc.rz(-angle / 2, target)

    

controlled_rz(qc3, 0, 1, 2 * np.arccos(a1))
controlled_rz(qc3, 0, 1, 2 * np.arccos(a1))
controlled_rz(qc3, 0, 1, 2 * np.arccos(a3))

qc3.cx(0, 1)




qc3.draw('mpl')


```

```python
sv = Statevector.from_instruction(qc3)

s = sv.draw('latex')
s
```
